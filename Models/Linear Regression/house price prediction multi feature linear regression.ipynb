### Group all the features as dependent features in X

```python
X = df.drop(['Y house price of unit area'], axis=1)
X
```

### Group feature(s) as independent features in y

```python
Y = df['Y house price of unit area']
Y
```

### Split the dataset into train and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

### Choose the model (Linear Regression)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

### Create an Estimator object

```python
model = LinearRegression()
```

### Train the model

```python
model.fit(X_train, y_train)
```

### Apply the model

```python
y_pred = model.predict(X_test)
y_pred
```

### Display the coefficients

```python
model.coef_
```

### Find how well the trained model did with testing data

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-squared Error or r2 score:', metrics.r2_score(y_test,y_pred))
```

### Plot Number of Convienience Stores Vs Price

```python
# prompt: Plot Number of Convienience Stores Vs Price, also print the total no of houses in the graph's x axis, the total no of houses present at each X4 value

plt.figure(figsize=(10,6))
sns.regplot(x=df['X4 number of convenience stores'], y=df['Y house price of unit area'], marker='x', color='.4', line_kws=dict(color='r'))
plt.xlabel("X4 Number of Convenience Stores")
plt.ylabel("Y House Price")
plt.title("Number of Convenience Stores vs Price")

# Calculate and display the total number of houses
total_houses = len(df)
plt.figtext(0.5, 0.01, f"Total Number of Houses: {total_houses}", ha="center", fontsize=10)

# Calculate and display the number of houses at each X4 value
x4_counts = df['X4 number of convenience stores'].value_counts().sort_index()
for x4_value, count in x4_counts.items():
    plt.text(x4_value, df['Y house price of unit area'].min(), f"{count}", ha='center', va='top', fontsize=8)

plt.show()
```
